```python
import BPt as bp
import numpy as np
import os
from os.path import dirname, abspath
```

```python
# Useful directories
main_dr = dirname(abspath(os.getcwd()))
data_dr = os.path.join(main_dr, 'data')

data = bp.read_pickle(os.path.join(data_dr, 'datasets', 'fs_thick_network.dataset'))
data
```

```python
# Define re-usable problem spec
ps = bp.ProblemSpec(target='IST_intelligence_total', n_jobs=8)
```

```python
data.plot(['age'])
```

```python
from BPt.default.pipelines import compare_pipe

# Pipeline is just identity + pre-defined comparison pipeline
loader = bp.Loader('identity')
pipe = bp.Pipeline([loader, compare_pipe])

# Evaluate & print summary
results = bp.evaluate(pipe, data, ps)
results.summary()
```
