```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import BPt as bp
import os
from os.path import dirname, abspath
```

```python
# Useful directories
main_dr = dirname(abspath(os.getcwd()))
data_dr = os.path.join(main_dr, 'data')

# Load in our pre-saved dataset
data = bp.read_pickle(os.path.join(data_dr, 'datasets', 'fs_rois.dataset'))
data
```

```python
ps = bp.ProblemSpec(target='BIS',
                    n_jobs=8,
                    random_state=1)

results = bp.evaluate('lgbm_pipe', data, ps, mute_warnings=True)


results
```

```python
p_results = results.permutation_importance(data, n_repeats=10000)
p_results['importances_mean'].mean().sort_values()
```
